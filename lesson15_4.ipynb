```python
import numpy as np

student = np.array([67, 78, 92, 74, 65])
student[4]
student
```

```python
import pandas as pd

s1 = pd.Series([67, 78, 92, 74, 65],index=['國文', '英文', '數學', '地理', '歷史'])
s1
```

```python
s1 + 3
```

```python
print(s1['國文']) #使用key-index
s1[0] #使用原生-index
```

```python
for i in s1:
    print(i)
```

```python
#numpy的統計 可套用在pandas
#利用numpy的function 作統計
print(s1,end="\n")
print(f"加總:", np.sum(s1))
print(f"平均:", np.mean(s1))
print(f"標準差:", np.std(s1))
print(f"中位數:", np.median(s1))
```

```python
#多組series 就可組成dataframe
import pandas as pd
pd.DataFrame({
    "徐小明":[78, 92, 76, 91, 72],
    "王大同":[72, 81, 92, 94, 69],
    "張一生":[81, 65, 89, 81, 72],
},index=['國文', '英文', '數學', '地理', '歷史'])
```

```python
#DataFrame 就是表格資料，可對應到csv
import numpy as np
import pandas as pd

data = np.random.randint(60,high=100,size=(50,5))
pd.DataFrame(data,
             index = range(1,51),
             columns=['國文', '英文', '數學', '地理', '歷史']
             )
```
