```python
#讀取csv檔成為2維的list
import csv
with open('a1.csv',encoding='utf-8',newline='') as file:
    csv_reader = csv.reader(file)
    next(csv_reader)
    #print(type(csv_reader))
    csv_list = list(csv_reader)
display(csv_list)
```

```python
for student in csv_list:
    if student[0] == '周家銘':
        print(student)
        break
```
